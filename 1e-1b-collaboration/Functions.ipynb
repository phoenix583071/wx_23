The function for saving a file.

```python
# NECESSARY IMPORTS
#---------------------------------------------------------------------------------------------

import numpy as np
import pandas as pd
import glob
import polars as pl
import datetime as dt

#---------------------------------------------------------------------------------------------
# CONVERTS BLM FILE TO NUMPY ARRAY, NOT NECESSARY JUST HERE FOR DATA
#---------------------------------------------------------------------------------------------

files = glob.glob("BLM_R5IM_Data/cycle" + '/*.csv')
selected_file = files[0]
input_data = pd.read_csv(selected_file)
dataframe = input_data.drop(columns = input_data.columns[0]).to_numpy()

#---------------------------------------------------------------------------------------------

def InitialiseBLMFile():
    correct_input = False
    while(correct_input == False):
        #file_directory = input('''Where would you like to store your file?
        #Example: C:\\Users\\izy65733\\documents''')
        global file_directory 
        file_directory = "C:\\Users\\izy65733\\documents\\SavedData\\"
        correct_input = True
    correct_input = False
    while(correct_input == False):
        global transfer_type 
        transfer_type = input('''What transfer type would you like to use?
        - numpy (recommended for 50hz standard BLM data)
        - polars
        (If you are reading use the type it was transferred in)''')
        if(transfer_type.lower() == "numpy"):
            correct_input = True
        if(transfer_type.lower() == "polars"):
            correct_input = True
            
def SaveBLMFile(data):
    global last_time_read
    last_time_read = '{date:%d_%m_%Y_%H_%M_%S_%f}'.format(date=dt.datetime.now())
    if(transfer_type == "numpy"):
        np.save(file_directory+last_time_read+".npy", data)
    if(transfer_type == "polars"):
        np.savetxt(file_directory+last_time_read+".csv",data, delimiter=",")

def ReadBLMFile():
    correct_input = False
    while(correct_input == False):     
            retrieve_last = input("Would you like to retrieve the last saved file?")
            if(retrieve_last.lower() == "yes"):
                correct_input = True
                read_directory = file_directory+last_time_read
            if(retrieve_last.lower() == "no"):
                correct_input = True   
                
    correct_input = False
    while(correct_input == False):
        if (retrieve_last.lower() == "no"):
            read_directory = input('''Which file would you like to retrieve?
            Example: C:\\Users\\izy65733\\documents\\SavedData\\26_05_2023_00_35_36_034042''')
        correct_input = True
        
    if(transfer_type == "numpy"):       
        new_dataframe = np.load(read_directory+".npy")
    if(transfer_type == "polars"):
        polars_array = pl.read_csv(read_directory+".csv")
        new_dataframe = polars_array.drop(columns = input_data.columns[0]).to_numpy()
    return new_dataframe

InitialiseBLMFile()
SaveBLMFile(dataframe)
new_dataframe = ReadBLMFile()
print(new_dataframe)
```
