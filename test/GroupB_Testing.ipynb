# Testing the time taken

Running every function 1000 times and timing how long this takes.

```python
import numpy as np
import statistics
row = 39
col = 2200
valueToFill = 1
arrayOf1s = np.full((row,col),valueToFill)
```

```python


import time as t
start = t.time()

class Class: 
          
    def sumOfNumbers(self, signal):
        return(sum(arrayOf1s[signal]))


    def meanOfNumbers(self, signal):
        return(statistics.mean(arrayOf1s[signal]))

    def time(self, times, startTime, endTime, dataPoints):
        return list(map(lambda t: (round(((t-startTime)/(endTime - startTime))*dataPoints)), times))

    def pairs(self, listOfTimes):
        return list(zip(listOfTimes, listOfTimes[1:]))

    def newestSum(self, signal, pairsList):
        return([sum(arrayOf1s[signal][start:end]) for start, end in pairsList])

    def newestMean(self, signal, pairsList):
        return ([statistics.mean(arrayOf1s[signal][start:end]) for start, end in pairsList])


for i in range(0, 1000):
    stats = Class()
    sumTest = stats.sumOfNumbers(0)
    meanTest = stats.meanOfNumbers(0)
    time1 = stats.time([-0.5, 0.0, 3.0, 4.6, 5.5], -0.5, 10.5, 2200)
    pair1 = stats.pairs(time1)
    sum1 = stats.newestSum(0, pair1)
    mean1 = stats.newestMean(0, pair1)




end = t.time()
print(end-start)
```
